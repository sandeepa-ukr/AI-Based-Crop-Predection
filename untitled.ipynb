##Combined Climate and Crop Data for Training

```python
def create_training_dataset(merged_df, target_element='Yield', target_unit='hg/ha'):
    """
    Create training data by merging climate features with crop targets
    """
    # Filter crop data for specific element and unit
    crop_target = crop_df[
        (crop_df['Element'] == target_element) &
        (crop_df['Unit'] == target_unit)
    ].copy()

    # Prepare climate data (keep only year and climate columns)
    climate_features = climate_df[['Year', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN',
                                   'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC', 'ANN']].copy()

    # Merge climate data with crop data on Year
    training_data = pd.merge(
        # climate_features,
        climate_df,
        crop_target[['Year', 'Item', 'Value']],
        on='Year',
        how='inner',
        suffixes=('_climate', '_crop')
    )

    # Rename target column
    training_data = training_data.rename(columns={'Value': 'target_yield'})

    return training_data.dropna()

# Create training dataset
train_data = create_training_dataset(merged_df, 'Yield', 'hg/ha')
print(f"Training data shape: {train_data.shape}")
print(f"Unique crops in training data: {train_data['Item'].nunique()}")
```

#Correlation-Based Feature Selection

```python
# Keep only numeric columns before correlation
numeric_data = train_data.select_dtypes(include=[np.number])

correlation_threshold = 0.1   # you may change to 0.1

corr_with_target = numeric_data.corr()['target_yield'].abs()
selected_corr_features = corr_with_target[corr_with_target > correlation_threshold].index.tolist()

print("Selected by Correlation (>", correlation_threshold, "):")
print(selected_corr_features)

numeric_data = numeric_data[selected_corr_features]

# Remove multicollinearity
corr_matrix = numeric_data.corr().abs()
upper_triangle = np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
upper_triangle_corr = corr_matrix.where(upper_triangle)

high_corr_threshold = 0.85
to_drop = [column for column in upper_triangle_corr.columns
           if any(upper_triangle_corr[column] > high_corr_threshold)]

print("Highly correlated features removed:", to_drop)

numeric_data = numeric_data.drop(columns=to_drop)

# Final numeric features (except target)
numeric_features = numeric_data.columns.tolist()
numeric_features.remove('target_yield')
```

```python
from sklearn.feature_selection import SelectKBest, f_regression
FTest_data = train_data.select_dtypes(include=[np.number])
X = FTest_data.drop(columns=['target_yield'])
y = FTest_data['target_yield']

k = 10  # choose how many features you want
selector = SelectKBest(score_func=f_regression, k=k)
X_new = selector.fit_transform(X, y)

selected_f_features = X.columns[selector.get_support()].tolist()

print("Top", k, "features selected using F-test:")
print(selected_f_features)

# Keep only these features
FTest_data = FTest_data[selected_f_features + ['target_yield']]
```

```python


# Find optimal number of clusters using elbow method
inertia = []
silhouette_scores = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X_scaled_df)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled_df, cluster_labels))

# Plot elbow curve and silhouette scores
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

ax1.plot(k_range, inertia, marker='o')
ax1.set_xlabel('Number of Clusters')
ax1.set_ylabel('Inertia')
ax1.set_title('Elbow Method')
ax1.grid(True)

ax2.plot(k_range, silhouette_scores, marker='o', color='orange')
ax2.set_xlabel('Number of Clusters')
ax2.set_ylabel('Silhouette Score')
ax2.set_title('Silhouette Scores')
ax2.grid(True)

plt.tight_layout()
plt.show()

# Choose optimal k (you can adjust based on the plots)
optimal_k = 4  # Change based on elbow/silhouette analysis
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X_scaled_df)

# Add cluster labels to original data
train_data['cluster'] = cluster_labels

print(f"Cluster distribution:")
print(train_data['cluster'].value_counts().sort_index())
```

```python
# 1. CLUSTERING: Group similar climate-crop patterns
def analyze_clusters():
    kmeans = KMeans(n_clusters=5, random_state=42)
    clusters = kmeans.fit_predict(X_scaled)

    # Analyze what each cluster represents
    cluster_profile = train_data.copy()
    cluster_profile['cluster'] = clusters

    print("Cluster characteristics:")
    for cluster_id in range(5):
        cluster_data = cluster_profile[cluster_profile['cluster'] == cluster_id]
        print(f"\nCluster {cluster_id} ({len(cluster_data)} samples):")
        if 'Item' in cluster_data.columns:
            print(f"  Common crops: {cluster_data['Item'].value_counts().head(3).to_dict()}")
        print(f"  Climate profile - Avg temp: {cluster_data['ANN'].mean():.1f}, "
              f"Rainfall: {cluster_data[['JAN', 'FEB']].mean().mean():.1f}")

# 2. DIMENSIONALITY REDUCTION: Simplify climate variables
def reduce_dimensions():
    pca = PCA(n_components=3)
    principal_components = pca.fit_transform(X_scaled)
    print(f"Variance explained: {pca.explained_variance_ratio_.sum():.2%}")

    # Identify which original features contribute most to each component
    feature_importance = pd.DataFrame(
        pca.components_,
        columns=X.columns,
        index=[f'PC{i+1}' for i in range(3)]
    )
    print("\nTop features for each principal component:")
    print(feature_importance.T.abs().idxmax(axis=1))

# 3. PATTERN DISCOVERY: Find climate-crop relationships
def discover_patterns():
    # Correlation between climate and crop types
    climate_crop_corr = train_data.groupby('Item')[[
        'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN',
        'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC'
    ]].mean()

    print("Climate patterns by crop type:")
    print(climate_crop_corr.head())

# 4. ANOMALY DETECTION: Find unusual data points
def detect_anomalies():
    iso_forest = IsolationForest(contamination=0.1, random_state=42)
    anomalies = iso_forest.fit_predict(X_scaled)

    anomalous_data = train_data[anomalies == -1]
    print(f"Detected {len(anomalous_data)} anomalous records")

    if 'Year' in anomalous_data.columns and 'Item' in anomalous_data.columns:
        print("Anomalous years and crops:")
        print(anomalous_data[['Year', 'Item']].head(10))
```

```python
# Start with PATTERN DISCOVERY + CLUSTERING combined
def comprehensive_analysis(train_data):
    print("=== PATTERN DISCOVERY ===")

    # 1. Climate-Crop Relationships
    if 'Item' in train_data.columns:
        crop_climate_corr = train_data.groupby('Item')[[
            'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN',
            'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC', 'ANN'
        ]].mean()

        print("Top climate patterns by crop:")
        print(crop_climate_corr.head(10))

    # 2. Seasonal Patterns
    print("\n=== SEASONAL ANALYSIS ===")
    seasonal_impact = train_data.groupby('Item')[[
        'winter_avg', 'summer_avg', 'monsoon_avg'
    ]].mean()
    print("Seasonal climate preferences by crop:")
    print(seasonal_impact.head(10))

    # 3. CLUSTERING Analysis
    print("\n=== CLUSTERING ANALYSIS ===")
    from sklearn.cluster import KMeans
    from sklearn.preprocessing import StandardScaler

    # Prepare features (exclude non-numerical)
    features = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN',
                'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC', 'ANN',
                'winter_avg', 'summer_avg', 'monsoon_avg']

    X = train_data[features].dropna()
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    kmeans = KMeans(n_clusters=4, random_state=42)
    clusters = kmeans.fit_predict(X_scaled)

    # Analyze clusters
    train_data_clustered = train_data.copy()
    train_data_clustered = train_data_clustered.loc[X.index]  # Align indices
    train_data_clustered['cluster'] = clusters

    print("Cluster Profiles:")
    for cluster_id in range(4):
        cluster_data = train_data_clustered[train_data_clustered['cluster'] == cluster_id]
        print(f"\n Cluster {cluster_id} ({len(cluster_data)} records):")
        if 'Item' in cluster_data.columns:
            top_crops = cluster_data['Item'].value_counts().head(3)
            print(f"   Common crops: {dict(top_crops)}")
        print(f"   Avg annual climate: {cluster_data['ANN'].mean():.1f}")
        print(f"   Climate variability: {cluster_data['climate_std'].mean():.1f}")

    return train_data_clustered

# Run the analysis
enhanced_data = comprehensive_analysis(train_data)
```
