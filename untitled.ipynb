#Final Training Data Preparation

```python
# Separate features and target using the correct column names
def get_X_y(train_data, target_col='target_yield'):
    """Prepare features (X) and target (y) for training"""
    # Exclude non-feature columns and the target
    exclude_cols = [
        'Year', 'Item', 'target_yield',  # These should not be features
        # Remove any other non-feature columns if they exist
        'Domain Code', 'Domain', 'Area Code (FAO)', 'Area',
        'Element Code', 'Element', 'Item Code (FAO)',
        'Year Code', 'Unit', 'Flag', 'Flag Description'
    ]

    # Only exclude columns that actually exist in the data
    exclude_cols = [col for col in exclude_cols if col in train_data.columns]

    feature_cols = [col for col in train_data.columns if col not in exclude_cols]

    X = train_data[feature_cols]
    y = train_data[target_col]

    return X, y, feature_cols

# Now use the function with your training data
X, y, feature_names = get_X_y(train_data, 'target_yield')

print(f"âœ… Features shape: {X.shape}")
print(f"âœ… Target shape: {y.shape}")
print(f"âœ… Feature names ({len(feature_names)}): {feature_names}")

# Check for any missing values
print(f"\nâœ… Missing values in features: {X.isnull().sum().sum()}")
print(f"âœ… Missing values in target: {y.isnull().sum()}")

# Display feature statistics
print(f"\nâœ… Feature statistics:")
print(X.describe())

# Display target statistics
print(f"\nâœ… Target statistics:")
print(f"Mean: {y.mean():.2f}")
print(f"Std:  {y.std():.2f}")
print(f"Min:  {y.min():.2f}")
print(f"Max:  {y.max():.2f}")
```
